```python
product_name=[]
product_price=[]
product_rating=[]
for i in range(2,7):
    url='https://www.flipkart.com/search?q=automatic+washing+machine&sid=j9e%2Cabm%2C8qx&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_10_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_10_na_na_na&as-pos=2&as-type=RECENT&suggestionId=automatic+washing+machine%7CWashing+Machines&requestId=238c182c-95c6-4263-9051-d0b83fc77610&as-searchtext=automatic+washing+machine&page='+str(i)
    r=requests.get(url)
    soup=BeautifulSoup(r.text,'lxml')
    np=soup.find('a',class_='_1LKTO3').get('href')
    cnp='https://www.flipkart.com'+np
    container=soup.find_all('div',class_='_2kHMtA')
    for i in container:
        name=i.find('div',class_='_4rR01T')
        try:
            names=name.text
        except:
            names='nan'
        product_name.append(names)   
    for i in container:
        name=i.find('div',class_='_30jeq3 _1_WHN1')
        try:
            prices=name.text
        except:
            prices=0
        product_price.append(prices)     
    for i in container:
        name=i.find('div',class_='_3LWZlK')
        try:
            ratings=name.text
        except:
            ratings=0
        product_rating.append(ratings) 
df=pd.DataFrame({'name':product_name,'price':product_price,'rating':product_rating})
```
