```python
productname=[]
productprice=[]
productrating=[]
productreview=[]
productdiscount=[]
productdelivery=[]
for i in range(2,7):
    url='https://www.flipkart.com/search?q=projector&sid=6bo%2Ctia%2C1hx&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=RECENT&suggestionId=projector%7CProjectors&requestId=531ee522-f510-471a-b041-0a61256bd89c&as-searchtext=projector&page='+str(i)
    r=requests.get(url)
    soup=BeautifulSoup(r.text,'lxml')
    np=soup.find('a',class_='_1LKTO3').get('href')
    cnp='https://www.flipkart.com'+np
    container=soup.find_all('div',class_='_4ddWXP')
    for i in container:
        name=i.find('a',class_='s1Q9rs')
        try:
            names=name.text.split(',')[0]
        except:
            names='nan' 
        productname.append(names)
    for i in container:
        price=i.find('div',class_='_30jeq3')
        try:
            prices=price.text
        except:
            prices=0
        productprice.append(prices)  
    for i in container:
        rating=i.find('div',class_='_3LWZlK')
        try:
            ratings=rating.text
        except:
            ratings=0
        productrating.append(ratings)
    for i in container:
        reviw=i.find('span',class_='_2_R_DZ')
        try:
            reviews=reviw.text.split('(')[1].split(')')[0]
        except:
            reviews=0
        productreview.append(reviews)
    for i in container:
        dis=i.find('div',class_='_3Ay6Sb')
        try:
            discount=dis.text.split(' ')[0]
        except:
            discount=0
        productdiscount.append(discount)
    for i in container:
        free=i.find('div',class_='_2Tpdn3')
        try:
            freedelivery=free.text
        except:
            freedelivery='nan'
        productdelivery.append(freedelivery)   
        

df=pd.DataFrame({'name':productname,'prices':productprice,'rating':productrating,'reviews':productreview,'dicounts':productdiscount,'fdelivery':productdelivery})        
    
```
